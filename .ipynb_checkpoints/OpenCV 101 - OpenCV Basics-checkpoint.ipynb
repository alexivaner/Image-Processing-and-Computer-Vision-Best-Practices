```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import imutils
```

```python
def plt_imshow(title,image):
    image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.title(title)
    plt.show()
```

# Load Image

```python
image_path="opencv-split-merge/opencv_logo.png"
image=cv2.imread(image_path)
```

```python
(h,w,c)=image.shape[:3]
```

```python
plt_imshow("Original Image",image)
```

# Getting Setting

```python
#Find center

(cX,cY)=(w//2,h//2)
```

```python
top_left=image[0:cY,0:cX]
top_right=image[0:cY,cX:w]
bottom_left=image[cY:h,0:cX]
bottom_right=image[cY:h,cX:w]
```

```python
plt_imshow("Top_Left",top_left)
plt_imshow("Top_Right",top_right)
plt_imshow("Bottom_Left",bottom_left)
plt_imshow("Bottom_Right",bottom_right)
```

# Translate Image

```python
#Using OpenCV

## Create translation matrices
M=np.float32(([1,0,50],[0,1,-50]))
image_tranlate=cv2.warpAffine(image,M,(w,h))
```

```python
plt_imshow("Translated",image_tranlate)
```

```python
# Using imutils
image_tranlate=imutils.translate(image,50,-50)
```

```python
plt_imshow("Translated Imutils",image_tranlate)
```

# OpenCV Drawing

```python
#Create canvas

canvas=np.zeros((h,w,c),dtype="uint8")
```

```python
image_draw=cv2.rectangle(canvas,(0,0),(100,100),(0,255,0),-1)
```

```python
plt_imshow("Rectangle",image_draw)
```

```python
image_draw=cv2.circle(canvas,(cX,cY),45,(255,0,0),3)
```

```python
plt_imshow("Circle",image_draw)
```

```python
image_draw=cv2.line(canvas,(0,0),(w,h),(0,0,255),5)
```

```python
plt_imshow("Line",image_draw)
```

# OpenCV Rotate

```python
#Using OpenCV

#Create Rotation Matrix

Mr=cv2.getRotationMatrix2D((cX,cY),30,1)
```

```python
image_rotate=cv2.warpAffine(image,Mr,(w,h))
```

```python
plt_imshow("image rotate",image_rotate)
```

```python
# using imutils

image_rotate2=imutils.rotate_bound(image,30)
```

```python
plt_imshow("image rotate 2",image_rotate2)
```

# Resizing

```python
newH=1000
newW=int((newH/h)*w)
```

```python
image_resize=cv2.resize(image,(newW,newH),cv2.INTER_LINEAR)
```

```python
plt_imshow("image_resize",image_resize)
```

```python
image_resize_2=imutils.resize(image,height=1000)
```

```python
plt_imshow("image_resize_2",image_resize_2)
```

# Image Flipping

```python
image_flip_vertical=cv2.flip(image,0)
image_flip_horizontal=cv2.flip(image,1)
image_flip_both=cv2.flip(image,-1)
```

```python
plt_imshow("image_flip_vertical",image_flip_vertical)
plt_imshow("image_flip_horizontal",image_flip_horizontal)
plt_imshow("image_flip_both",image_flip_both)
```

# Cropping MaskingBitwise

```python
# Cropping using numpy slicing

crop_1=image[120:500,200:400]
```

```python
plt_imshow("crop_1",crop_1)
```

```python
# Cropping using masking
mask = np.zeros(image.shape[:2], dtype="uint8")
```

```python
masking=cv2.circle(mask,(cX,cY),5,30,255,-1)
```

```python
plt_imshow("masking-crop",masking)
```

```python
masked = cv2.bitwise_and(image, image, mask=masking)
```

```python
plt_imshow("masked",masked)
```

# Split and Merge

```python
(B,G,R)=cv2.split(image)
```

```python
plt_imshow("red",R)
plt_imshow("green",G)
plt_imshow("blue",B)
```

```python
merge=cv2.merge((R,G,B))
```

```python
plt_imshow("merged",merge)
```
